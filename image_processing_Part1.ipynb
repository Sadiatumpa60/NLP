#Colors with Numpy for RGB SCALE

```python
red=img[:,:,0]
green=img[:,:,1]
blue=img[:,:,2]
fig = plt.figure(figsize=(10,10))

ax1 = fig.add_subplot(2,2,1)
ax1.imshow(img)
ax1.title.set_text('Original')

ax2 = fig.add_subplot(2,2,2)
ax2.imshow(red)
ax2.title.set_text('Red')

ax3 = fig.add_subplot(2,2,3)
ax3.imshow(green)
ax3.title.set_text('Green')

ax2 = fig.add_subplot(2,2,4)
ax2.imshow(blue)
ax2.title.set_text('Blue')
```

#Colors with Numpy for Gray SCALE

```python
fig = plt.figure(figsize=(10,10))

ax1 = fig.add_subplot(2,2,1)
ax1.imshow(img)
ax1.title.set_text('Original')

ax2 = fig.add_subplot(2,2,2)
ax2.imshow(red,cmap="gray")
ax2.title.set_text('Red')

ax3 = fig.add_subplot(2,2,3)
ax3.imshow(green,cmap="gray")
ax3.title.set_text('Green')

ax2 = fig.add_subplot(2,2,4)
ax2.imshow(blue,cmap="gray")
ax2.title.set_text('Blue')
```
